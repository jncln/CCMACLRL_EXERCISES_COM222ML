```python
import pandas as pd
import seaborn as sns

from matplotlib import pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import cross_val_score
```

## Dataset File

```python
dataset_url = 'https://github.com/robitussin/CCMACLRL_EXERCISES/blob/main/datasets/loan_approval/train.csv?raw=true'
df = pd.read_csv(dataset_url)
```

## Test File

```python
test_url = 'https://github.com/robitussin/CCMACLRL_EXERCISES/blob/main/datasets/loan_approval/test.csv?raw=true'
dt=pd.read_csv(test_url)
```

# Preprocess Data

```python
categorical_cols = df.select_dtypes(include=['object']).columns
df = pd.get_dummies(df, columns=categorical_cols)

dt = pd.get_dummies(dt, columns=categorical_cols)

dt = dt.reindex(columns=df.columns, fill_value=0)
```

```python
X = df.drop('loan_status', axis=1)
y = df['loan_status']

dt = dt[X.columns]
```

## 1. Train a KNN Classifier

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X, y)
```

- Perform cross validation

```python
knn_scores = cross_val_score(knn, X, y, cv=5, scoring='accuracy')
print("KNN Cross-Validation Accuracy:", knn_scores.mean())
```

## 2. Train a Logistic Regression Classifier

```python
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X, y)
```

- Perform cross validation

```python
log_reg_scores = cross_val_score(log_reg, X, y, cv=5, scoring='accuracy')
print("Logistic Regression Cross-Validation Accuracy:", log_reg_scores.mean())
```

## 3. Train a Naive Bayes Classifier

```python
nb = GaussianNB()
nb.fit(X, y)
```

- Perform cross validation

```python
nb_scores = cross_val_score(nb, X, y, cv=5, scoring='accuracy')
print("Naive Bayes Cross-Validation Accuracy:", nb_scores.mean())
```

## 4. Train a SVM Classifier

```python
svm = SVC()
svm.fit(X, y)
```

- Perform cross validation

```python
svm_scores = cross_val_score(svm, X, y, cv=5, scoring='accuracy')
print("SVM Cross-Validation Accuracy:", svm_scores.mean())
```

## 5. Train a Decision Tree Classifier

```python
tree = DecisionTreeClassifier()
tree.fit(X, y)
```

- Perform cross validation

```python
tree_scores = cross_val_score(tree, X, y, cv=5, scoring='accuracy')
print("Decision Tree Cross-Validation Accuracy:", tree_scores.mean())
```

## 6. Train a Random Forest Classifier

```python
rf = RandomForestClassifier(n_estimators=100)
rf.fit(X, y)

rf_scores = cross_val_score(rf, X, y, cv=5, scoring='accuracy')
print("Random Forest Cross-Validation Accuracy:", rf_scores.mean())
```

## 7. Compare all the performance of all classification models

```python
print("Model Performance Comparison:")
print(f"KNN Accuracy: {knn_scores.mean()}")
print(f"Logistic Regression Accuracy: {log_reg_scores.mean()}")
print(f"Naive Bayes Accuracy: {nb_scores.mean()}")
print(f"SVM Accuracy: {svm_scores.mean()}")
print(f"Decision Tree Accuracy: {tree_scores.mean()}")
print(f"Random Forest Accuracy: {rf_scores.mean()}")
```
